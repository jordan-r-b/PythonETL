```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("Distro_Detail_v1.xlsx")
```

```python
df2 = df[df['Account'].str.contains("Investor ")]
```

```python
df_rep = pd.read_excel("Rep_Lookup.xlsx", header = None, names = ['A','B','C'])
```

```python
df_rep = df_rep.drop(['A'], axis=1)
```

```python
df2 = df2.merge(df_rep, how='left', left_on = 'Account', right_on = 'B')
```

```python
df2['Month'] = df2['Distribution date'].dt.month
df2['Quarter'] = df2['Distribution date'].dt.quarter
```

```python
df_pivot = pd.pivot_table(df2,index=["C"], values=["Gross distributions"], 
columns = ["Quarter","Month"], aggfunc=[np.sum],fill_value=0,margins=True)
```

```python
plt.figure(figsize=(20,6))
sns.swarmplot(x='C', 
              y='Gross distributions', 
              data=df2, 
              hue='Quarter', 
              split=True)
```
